```python
from PIL import Image
import IPython.display as display

# Python kétdimenziós lista indexelése:
# A lista esetében a formátum: lista[sor][oszlop]
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
# A 2. sor (index 1) 3. oszlopának (index 2) értéke:
print("Python lista, matrix[1][2]:", matrix[1][2])  # Eredmény: 6

# Hozzunk létre egy 3x3 pixeles képet fekete háttérrel:
image = Image.new("RGB", (3, 3), (0,0,200))
pixels = image.load()

# Pillow esetében a pixel indexelése: pixels[x, y],
# ahol x az oszlop (vízszintes) és y a sor (függőleges).
# Tehát a (2, 1) azt jelenti: 3. oszlop (x=2) és 2. sor (y=1).
pixels[2, 1] = (255, 255, 255)  # Beállítjuk a pixelt fehérre

print("Pillow kép, pixels[2, 1]:", pixels[2, 1])  # Eredmény: (255, 255, 255)

image = image.resize((500,500),resample=0)

# Kép megjelenítése
display.display(image)
```

```python
from PIL import Image

# Kép betöltése (cseréld ki az elérési utat)
image_path = "/content/tolgy.png"  # Fájl elérési útja
image = Image.open(image_path)

# Kép méretének lekérése
width, height = image.size

# Keret méretei
border_size = 20  # Fekete keret vastagsága
inner_border = 5  # Fehér csík vastagsága

# Új kép létrehozása, ami nagyobb, hogy elférjen a keret
new_width = width + 2 * (border_size + inner_border)
new_height = height + 2 * (border_size + inner_border)
new_image = Image.new("RGB", (new_width, new_height), "black")

# A fehér belső csík hozzáadása
for x in range(border_size, new_width - border_size):
    for y in range(border_size, new_height - border_size):
        if (x < border_size + inner_border or x >= new_width - (border_size + inner_border) or
            y < border_size + inner_border or y >= new_height - (border_size + inner_border)):
            new_image.putpixel((x, y), (255, 255, 255))  # Fehér belső keret

# Az eredeti kép beillesztése a középre
new_image.paste(image, (border_size + inner_border, border_size + inner_border))

display.display(new_image)
```

```python
from PIL import Image

# Kép betöltése
img = Image.open("/content/tolgy.png")
img = img.convert("RGB")  # RGB módba konvertálás
pixels = img.load()  # Pixelek tömbként való elérése

width, height = img.size
border_size = 10  # Szélek vastagsága
darken_factor = 0.5  # 50%-kal sötétebbé tesszük

for x in range(width):
    for y in range(height):
        # Ha a pixel a széleken van
        if x < border_size or x >= width - border_size or y < border_size or y >= height - border_size:
            r, g, b = pixels[x, y]  # Eredeti szín kiolvasása
            r = int(r * darken_factor)
            g = int(g * darken_factor)
            b = int(b * darken_factor)
            pixels[x, y] = (r, g, b)  # Új szín beállítása

display.display(img)

```

```python
from PIL import Image, ImageDraw

# Kép betöltése
img = Image.open("/content/tolgy.png")
img = img.convert("RGB")  # RGB módba konvertálás
pixels = img.load()  # Pixelek tömbként való elérése

width, height = img.size
border_size = 10  # Sötétített szélek vastagsága
darken_factor = 0.5  # 50%-kal sötétebbé tesszük
white_border_size = 2  # Fehér keret vastagsága

# Sötétített keret létrehozása
for x in range(width):
    for y in range(height):
        # Ha a pixel a széleken van (sötétítés)
        if x < border_size or x >= width - border_size or y < border_size or y >= height - border_size:
            r, g, b = pixels[x, y]  # Eredeti szín kiolvasása
            r = int(r * darken_factor)
            g = int(g * darken_factor)
            b = int(b * darken_factor)
            pixels[x, y] = (r, g, b)  # Új szín beállítása

# Fehér keret hozzáadása a sötétített rész belső oldalán
for x in range(border_size, width - border_size):
    for y in range(border_size, height - border_size):
        # Ha a pixel a fehér keret sávjában van
        if (border_size <= x < border_size + white_border_size or
            width - border_size - white_border_size <= x < width - border_size or
            border_size <= y < border_size + white_border_size or
            height - border_size - white_border_size <= y < height - border_size):
            pixels[x, y] = (255, 255, 255)  # Fehér pixel

# Kép megjelenítése
display.display(img)
```

```python
from PIL import Image

# Kép betöltése
img = Image.open("/content/tolgy.png")
img = img.convert("RGB")  # RGB módba konvertálás
pixels = img.load()  # Pixelek tömbként való elérése

width, height = img.size

# Szépia filter alkalmazása
for x in range(width):
    for y in range(height):
        r, g, b = pixels[x, y]  # Eredeti szín kiolvasása

        # Új szépia színek számítása
        new_r = int(0.393 * r + 0.769 * g + 0.189 * b)
        new_g = int(0.349 * r + 0.686 * g + 0.168 * b)
        new_b = int(0.272 * r + 0.534 * g + 0.131 * b)

        # 255 feletti értékeket levágunk
        pixels[x, y] = (min(255, new_r), min(255, new_g), min(255, new_b))

# Kép megjelenítése
display.display(img)
```

```python
from PIL import Image

img = Image.open("tolgy.png")
pixels = img.load()

width, height = img.size

for x in range(width):
    for y in range(height):
        r, g, b = pixels[x, y]
        if y < height // 2:  # Csak a kép felső felét alakítjuk át
            gray = int(0.3 * r + 0.59 * g + 0.11 * b)  # Szürkeárnyalat
            pixels[x, y] = (gray, gray, gray)

display.display(img)
```
